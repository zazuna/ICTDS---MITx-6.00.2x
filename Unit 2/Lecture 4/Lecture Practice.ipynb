```python
import pylab as plt
mySamples = []
myLinear = []
myQuadratic = []
myCubic = []
myExponential = []

for i in range(30):
    mySamples.append(i)
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myExponential.append(1.5**i)
    
plt.figure('lin')
plt.xlabel('Samples Points')
plt.ylabel('Linear Function')
plt.title('Linear')
plt.clf()
plt.plot(mySamples, myLinear)
plt.figure('quad')
plt.xlabel('Samples Points')
plt.ylabel('Quadratic Function')
plt.title('Quadratic')
plt.plot(mySamples, myQuadratic)
plt.figure('cub')
plt.xlabel('Samples Points')
plt.ylabel('Cubic Function')
plt.title('Cubic')
plt.plot(mySamples, myCubic)
plt.figure('exp')
plt.xlabel('Samples Points')
plt.ylabel('Exponential Function')
plt.title('Exponential')
plt.plot(mySamples, myExponential)
```

```python
plt.figure('lin quad')
plt.clf()
plt.title('Linear vs Quadratic')
plt.plot(mySamples, myLinear, 'b-' , label='Linear', linewidth=1)
plt.plot(mySamples, myQuadratic, 'r--', label='Quadratic')
plt.legend(loc='best')
plt.figure('cub exp')
plt.clf()
plt.title('Cubic vs Exponential')
plt.plot(mySamples, myCubic, 'r^', label='Cubic')
plt.plot(mySamples, myExponential, 'go', label='Exponential')
plt.legend(loc='best')
```

```python
plt.figure('lin quad')
plt.clf()
plt.title('Linear vs Quadratic')
plt.subplot(2,1,1)
plt.ylim([0,900])
plt.plot(mySamples, myLinear, 'b-' , label='Linear', linewidth=1)
plt.subplot(2,1,2)
plt.plot(mySamples, myQuadratic, 'r--', label='Quadratic')

plt.figure('cub exp')
plt.clf()
plt.title('Cubic vs Exponential')
plt.subplot(1,2,1)
# plt.ylim([0,140000])
plt.yscale('log')
plt.plot(mySamples, myCubic, 'r-', label='Cubic', linewidth=1)
plt.subplot(1,2,2)
# plt.ylim([0,140000])
plt.yscale('log')
plt.plot(mySamples, myExponential, 'g--', label='Exponential', linewidth=1)

```

```python
def retire(monthly, rate, terms):
    savings = [0]
    base = [0]
    mRate = rate/12
    for i in range(terms):
        base += [i]
        savings += [savings[-1] * (1 + mRate) + monthly]
    return base, savings

def displayRetireWMonthlies(monthlies, rate, terms):
    plt.figure('retireMonth')
    plt.clf()
    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, '--', label='retire:' + str(monthly))
        plt.legend(loc='best')
        print(max(yvals))
        
def displayRetireWRates(monthlies, rates, terms):
    print('-' * len('With Diff Rates'), '\nWith Diff Rates')
    plt.figure('retireRate')
    plt.clf()
    for rate in rates:
        xvals, yvals = retire(monthly, rate, terms)
        plt.plot(xvals, yvals, '--', label='retire:' + str(monthly) + ' : ' + str(rate))
        plt.legend(loc='best')
        print(max(yvals))


def displayRetireWMonthsAndRates(monthlies, rates, terms):
    plt.figure('retireMonthRates', figsize=(10,10))
    plt.clf()
    # plt.xlim([30*12, 40*12])
    for monthly in monthlies:
        for rate in rates:
            xvals, yvals = retire(monthly, rate, terms)
            plt.plot(xvals, yvals, '--', label='retire: ' + str(monthly) + ' : ' + str(rate))
            plt.legend(loc='upper left')

def displayRetireWMonthsAndRatesLabels(monthlies, rates, terms):
    plt.figure('retireMonthRates', figsize=(10,10))
    plt.clf()
    plt.xlim([30*12, 40*12])
    monthLabels = ['r', 'b', 'g', 'k']
    rateLabels = ['-', '--', 'o']
    for i in range(len(monthlies)):
        monthly = monthlies[i]
        monthLabel = monthLabels[i%len(monthLabels)]
        for j in range(len(rates)):
            rate = rates[j]
            rateLabel = rateLabels[j%len(rateLabels)]
            xvals, yvals = retire(monthly, rate, terms)
            plt.plot(xvals, yvals, monthLabel + rateLabel, label='retire: ' + str(monthly) + ' : ' + str(rate), linewidth=1)
            plt.legend(loc='upper left')

monthlies = [500, 600, 700, 800, 900, 1000, 1100]
rate = 0.05
terms = 12 * 40
displayRetireWMonthlies(monthlies, rate, terms)
monthly = 800
rates = [0.03, 0.05, 0.07]
displayRetireWRates(monthly, rates, terms)
displayRetireWMonthsAndRates(monthlies, rates, terms)
displayRetireWMonthsAndRatesLabels(monthlies, rates, terms)
```
